```python
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib qt
plt.rcParams.update({'font.size': 20})
plt.rcParams['figure.figsize'] = [15, 12]
```

```python
import numpy as np
#time is coloumn 0
tme = np.genfromtxt("Cooldown21_2_11_2023_LP_Control.dat",usecols = 0)

#pot temperature
temp = np.genfromtxt("Cooldown21_2_11_2023_LP_Control.dat",usecols = 5)

#resistance of the sample is coloumn 9
res = np.genfromtxt("Cooldown21_2_11_2023_LP_Control.dat",usecols = 9)

#flag temperature
ftemp = np.genfromtxt("Cooldown21_2_11_2023_LP_Control.dat",usecols = 11)

for i in range(len(tme)):
    res[i] = 10000000*res[i]/25

plt.grid(linestyle = ':')
plt.ylabel('Resistance(Ohms)')
plt.xlabel('Time(s)')

plt.title('Resistance vs. Time')

plt.plot(tme,res,label = 'Hold 5, 1.745K')

plt.legend()
plt.show()
```

```python
res1 = []
tme1 = []
temp1 = []
ftemp1 = []
for i in range(len(tme)):
    if tme[i] > 6300 and tme[i] < 9200:
        tme1.append(tme[i])
        res1.append(res[i])
        temp1.append(temp[i])
        ftemp1.append(ftemp[i])
```

```python
plt.grid(linestyle = ':')
plt.ylabel('Resistance(Ohms)')
plt.xlabel('Time(s)')

plt.title('Resistance vs. Time')

plt.plot(tme1,res1,label = 'Hold 5, 1.745K, Std Dev: ' + str(np.std(res1)) + 'Ohm')

plt.legend()
plt.show()
```

```python
res1 = []
tme1 = []
temp1 = []
ftemp1 = []
for i in range(len(tme)):
    if tme[i] > 4463 and tme[i] < 4470:
        tme1.append(tme[i])
        res1.append(res[i])
        temp1.append(temp[i])
        ftemp1.append(ftemp[i])
```

```python
plt.title('Resistance vs. Pot Temperature (pot controlled at 1.72K and 1.745K)')
plt.plot(temp1,res1,'o',label = 'Hold 5, 1.745K')

plt.ylabel('Resistance(Ohms)')
plt.xlabel('Temperature(K)')
plt.grid(linestyle = ':')
a, b = np.polyfit(temp1, res1, 1)

tempf1 = []

for i in range(len(temp1)):
    tempf1.append(a*temp1[i] + b)
    
plt.plot(temp1, tempf1,label = 'Slope: ' + str(a) + ' Intercept : ' + str(b))

plt.legend()
plt.show()
```
